# MapLibre Basemap Control Example

This notebook demonstrates how to use the basemap control with MapLibre GL JS maps in AnyMap. The basemap control allows users to switch between different basemap providers using an interactive control widget. It uses the [maplibre-gl-basemaps](https://github.com/ka7eh/maplibre-gl-basemaps) plugin.

```python
from anymap import MapLibreMap
from anymap.basemaps import available_basemaps
```

## Basic Basemap Control

Create a map with a basic basemap control using default basemaps. The control appears as a button that expands to show basemap options when clicked or hovered over:

```python
# Create a MapLibre map centered on San Francisco
m = MapLibreMap(
    center=[-122.4194, 37.7749],
    zoom=12,
    # style="liberty",
    height="600px",
)

# Add basemap control with default basemaps
# Default basemaps: OpenStreetMap.Mapnik, Esri.WorldImagery, CartoDB.DarkMatter, CartoDB.Positron
m.add_basemap_control(position="top-left")

m
```

## Custom Basemap Selection

You can customize which basemaps appear in the control by providing a list of basemap names:

```python
# Create a map with custom basemap selection
m2 = MapLibreMap(
    center=[2.3522, 48.8566], zoom=11, style="liberty", height="600px"  # Paris
)

# Add control with custom basemaps
m2.add_basemap_control(
    position="top-left",
    basemaps=[
        "OpenStreetMap.Mapnik",
        "Esri.WorldImagery",
        "Esri.WorldTopoMap",
        "CartoDB.DarkMatter",
        "CartoDB.Positron",
        "OpenTopoMap",
    ],
    initial_basemap="OpenStreetMap.Mapnik",
)

m2
```

## Custom Labels

You can provide custom display labels for basemaps to make them more user-friendly:

```python
# Create a map with custom basemap labels
m3 = MapLibreMap(
    center=[-74.0059, 40.7128],  # New York City
    zoom=10,
    style="liberty",
    height="600px",
)

# Add control with custom labels
m3.add_basemap_control(
    position="top-right",
    basemaps=[
        "OpenStreetMap.Mapnik",
        "Esri.WorldImagery",
        "CartoDB.DarkMatter",
        "CartoDB.Positron",
    ],
    labels={
        "OpenStreetMap.Mapnik": "Street Map",
        "Esri.WorldImagery": "Satellite",
        "CartoDB.DarkMatter": "Dark",
        "CartoDB.Positron": "Light",
    },
    initial_basemap="OpenStreetMap.Mapnik",
)

m3
```

## Different Expand Directions

The basemap control can expand in different directions: 'up', 'down', 'left', or 'right':

```python
# Create a map with control that expands upward
m4 = MapLibreMap(
    center=[139.6917, 35.6895], zoom=11, style="liberty", height="600px"  # Tokyo
)

# Add control at bottom-right that expands upward
m4.add_basemap_control(
    position="bottom-right",
    basemaps=[
        "OpenStreetMap.Mapnik",
        "Esri.WorldImagery",
        "Esri.WorldStreetMap",
        "CartoDB.Voyager",
    ],
    expand_direction="up",
    labels={
        "OpenStreetMap.Mapnik": "OSM",
        "Esri.WorldImagery": "Satellite",
        "Esri.WorldStreetMap": "Esri Streets",
        "CartoDB.Voyager": "Voyager",
    },
)

m4
```

## Explore Available Basemaps

You can explore all available basemaps from the xyzservices library:

```python
# List all available basemap names
basemap_names = list(available_basemaps.keys())
print(f"Total available basemaps: {len(basemap_names)}")
print("\nFirst 20 basemaps:")
for name in basemap_names[:20]:
    print(f"  - {name}")
```

## Combining with Other Controls

The basemap control works well alongside other map controls:

```python
# Create a comprehensive map with multiple controls
m5 = MapLibreMap(
    center=[13.4050, 52.5200], zoom=11, style="liberty", height="600px"  # Berlin
)

# Add basemap control
m5.add_basemap_control(
    position="top-left",
    basemaps=[
        "OpenStreetMap.Mapnik",
        "Esri.WorldImagery",
        "CartoDB.DarkMatter",
        "Stamen.Terrain",
    ],
    labels={
        "OpenStreetMap.Mapnik": "Streets",
        "Esri.WorldImagery": "Satellite",
        "CartoDB.DarkMatter": "Dark Theme",
        "Stamen.Terrain": "Terrain",
    },
)

# Add other controls
m5.add_control("navigation", "top-right")
m5.add_control("fullscreen", "top-right")
m5.add_control("scale", "bottom-left")

m5
```

## Export to HTML

The basemap control is preserved when exporting the map to HTML:

```python
# Export map with basemap control to HTML
m3.to_html("basemap_control_map.html")
print("Map with basemap control exported to basemap_control_map.html")
```

## Different Positions

You can place the basemap control in any corner of the map:

```python
# Create a map with control in bottom-left
m6 = MapLibreMap(
    center=[-0.1276, 51.5074], zoom=11, style="liberty", height="600px"  # London
)

m6.add_basemap_control(
    position="bottom-left",
    basemaps=[
        "OpenStreetMap.Mapnik",
        "Esri.WorldImagery",
        "Esri.NatGeoWorldMap",
        "CartoDB.Voyager",
    ],
    expand_direction="right",
)

m6
```
