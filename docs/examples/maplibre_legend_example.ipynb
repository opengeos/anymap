# MapLibre Legend Control Example

This notebook demonstrates how to use the MapLibre GL Legend plugin with anymap. The legend control automatically generates legend entries from MapLibre layer styles.

```python
from anymap import MapLibreMap
```

## Basic Legend Example

Let's create a map with several layers and add a legend control that shows all layers.

```python
# Create a map centered on the United States
m = MapLibreMap(
    center=[-100, 40],
    zoom=3.5,
    height="600px",
)

# Add some example GeoJSON layers with different styles

# Layer 1: Cities (red circles)
cities_data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "geometry": {"type": "Point", "coordinates": [-74.0060, 40.7128]},
            "properties": {"name": "New York", "population": 8000000},
        },
        {
            "type": "Feature",
            "geometry": {"type": "Point", "coordinates": [-118.2437, 34.0522]},
            "properties": {"name": "Los Angeles", "population": 4000000},
        },
        {
            "type": "Feature",
            "geometry": {"type": "Point", "coordinates": [-87.6298, 41.8781]},
            "properties": {"name": "Chicago", "population": 2700000},
        },
    ],
}

m.add_geojson_layer(
    layer_id="cities",
    geojson_data=cities_data,
    layer_type="circle",
    paint={
        "circle-radius": 10,
        "circle-color": "#ff0000",
        "circle-stroke-width": 2,
        "circle-stroke-color": "#ffffff",
    },
)

# Layer 2: Parks (green polygons)
parks_data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "geometry": {
                "type": "Polygon",
                "coordinates": [
                    [
                        [-95.5, 39.5],
                        [-94.5, 39.5],
                        [-94.5, 40.5],
                        [-95.5, 40.5],
                        [-95.5, 39.5],
                    ]
                ],
            },
            "properties": {"name": "Sample Park"},
        }
    ],
}

m.add_geojson_layer(
    layer_id="parks",
    geojson_data=parks_data,
    layer_type="fill",
    paint={"fill-color": "#00ff00", "fill-opacity": 0.5},
)

# Layer 3: Roads (blue lines)
roads_data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "geometry": {
                "type": "LineString",
                "coordinates": [
                    [-100, 38],
                    [-98, 39],
                    [-96, 40],
                ],
            },
            "properties": {"name": "Highway 1"},
        }
    ],
}

m.add_geojson_layer(
    layer_id="roads",
    geojson_data=roads_data,
    layer_type="line",
    paint={"line-color": "#0000ff", "line-width": 3},
)

# Add legend control showing all layers
m.add_legend_control(
    position="top-right",
    targets={},  # Empty dict means include all layers
    title="Map Layers",
    show_default=True,
)

m
```

## Legend with Specific Layers

You can also specify which layers to include in the legend by providing a dictionary mapping layer IDs to display names.

```python
# Create another map
m2 = MapLibreMap(
    center=[-100, 40],
    zoom=3.5,
    height="600px",
)

# Add the same layers
m2.add_geojson_layer(
    layer_id="cities",
    geojson_data=cities_data,
    layer_type="circle",
    paint={
        "circle-radius": 10,
        "circle-color": "#ff0000",
        "circle-stroke-width": 2,
        "circle-stroke-color": "#ffffff",
    },
)

m2.add_geojson_layer(
    layer_id="parks",
    geojson_data=parks_data,
    layer_type="fill",
    paint={"fill-color": "#00ff00", "fill-opacity": 0.5},
)

m2.add_geojson_layer(
    layer_id="roads",
    geojson_data=roads_data,
    layer_type="line",
    paint={"line-color": "#0000ff", "line-width": 3},
)

# Add legend control with custom layer names and only specific layers
m2.add_legend_control(
    position="top-right",
    targets={
        "cities": "Major Cities",
        "parks": "National Parks",
    },  # Only include cities and parks in legend
    title="Selected Layers",
    show_default=True,
)

m2
```

## Legend with Different Options

The legend control supports various customization options.

```python
# Create a map with custom legend settings
m3 = MapLibreMap(
    center=[-100, 40],
    zoom=3.5,
    height="600px",
)

# Add multiple layers with different colors
for i, color in enumerate(["#ff0000", "#00ff00", "#0000ff", "#ffff00", "#ff00ff"]):
    layer_data = {
        "type": "FeatureCollection",
        "features": [
            {
                "type": "Feature",
                "geometry": {
                    "type": "Point",
                    "coordinates": [-100 + i * 2, 40 + i],
                },
                "properties": {"name": f"Point {i}"},
            }
        ],
    }

    m3.add_geojson_layer(
        layer_id=f"layer_{i}",
        geojson_data=layer_data,
        layer_type="circle",
        paint={
            "circle-radius": 8,
            "circle-color": color,
        },
    )

# Add legend with custom options
m3.add_legend_control(
    position="bottom-right",
    targets={},  # All layers
    title="Custom Legend",
    show_default=False,  # Start with legend hidden
    show_checkbox=True,  # Show checkboxes to toggle layer visibility
    reverse_order=False,  # Don't reverse the layer order
    only_rendered=True,  # Only show currently visible layers
)

m3
```

## Export to HTML

The legend control also works when exporting maps to HTML files.

```python
# Export the map to HTML
m.to_html("maplibre_with_legend.html", title="MapLibre Map with Legend")
print("Map exported to maplibre_with_legend.html")
```

## Advanced Example: Raster Layer with Legend

The legend control also works with raster layers and other MapLibre layer types.

```python
# Create a map with a tile layer
m4 = MapLibreMap(
    center=[-122.4194, 37.7749],  # San Francisco
    zoom=12,
    height="600px",
)

# Add a tile layer
m4.add_tile_layer(
    layer_id="satellite",
    source_url="https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}",
    attribution="Google",
)

# Add some vector layers on top
poi_data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "geometry": {"type": "Point", "coordinates": [-122.4194, 37.7749]},
            "properties": {"name": "City Hall"},
        },
        {
            "type": "Feature",
            "geometry": {"type": "Point", "coordinates": [-122.4183, 37.7793]},
            "properties": {"name": "Golden Gate Park"},
        },
    ],
}

m4.add_geojson_layer(
    layer_id="poi",
    geojson_data=poi_data,
    layer_type="circle",
    paint={
        "circle-radius": 8,
        "circle-color": "#ff6b6b",
        "circle-stroke-width": 2,
        "circle-stroke-color": "#ffffff",
    },
)

# Add legend with custom names
m4.add_legend_control(
    position="top-right",
    targets={
        "satellite": "Satellite Imagery",
        "poi": "Points of Interest",
    },
    title="San Francisco Layers",
)

m4
```
