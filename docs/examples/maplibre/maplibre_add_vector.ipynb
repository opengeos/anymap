```python
import anymap
```

```python
m = anymap.Map(style="liberty")
m.add_basemap("OpenTopoMap", before_id=m.first_symbol_layer_id)
m.add_basemap("Esri.WorldImagery", before_id=m.first_symbol_layer_id)
filepath = "https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip_train_buildings.geojson"
m.add_vector(filepath, opacity=0.5, name="buildings")
m.add_popup("buildings", trigger="hover")
m
```
