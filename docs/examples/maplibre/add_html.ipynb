# Add HTML control

```python
from anymap.maplibre import MapLibreMap

# Create map
m = MapLibreMap(center=[-74.0, 40.7], zoom=10)

# Add HTML control
m.add_html(
    html="<h3>Custom Info</h3><p>Some content here</p>",
    bg_color="lightblue",
    position="top-right",
    control_id="my_custom_control",
)
m
```

```python
# Update the HTML
m.update_html(
    control_id="my_custom_control",
    html="<h3>Updated Info</h3><p>New content</p>",
    bg_color="lightgreen",
)
```

```python
# Remove the HTML control
m.remove_html(control_id="my_custom_control")
```

```python
m.add_html(html="<div>Auto-generated key</div>", position="bottom-right")
```
