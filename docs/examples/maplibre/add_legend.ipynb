# Add legends to the map

```python
import anymap
```

```python
m = anymap.Map()
m.add_legend(
    title="Land Cover Type",
    builtin_legend="NLCD",
    header_color="linear-gradient(135deg,#444,#888)",
    collapsed=False,
    position="top-right",
    icon="ðŸŒ³",
    panel_max_height=380,
)
m
```

```python
m = anymap.Map()
m.add_legend(
    title="My Legend",
    labels=["Short", "Very Long Label Name That Will Expand Width"],
    colors=["#ff0000", "#00ff00"],
    collapsed=True,
)

m.add_legend(
    title="Fixed Width",
    labels=["Item 1", "Item 2"],
    colors=["#ff0000", "#00ff00"],
    responsive=True,
    collapsed=False,
)
m
```
